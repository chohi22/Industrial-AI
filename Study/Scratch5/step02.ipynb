### 최대 가능도 추정

```python
import os
import numpy as np
import matplotlib.pyplot as plt

#path = os.path.join(os.path.dirname(__file__), 'height.txt')
path = os.path.join(os.getcwd(), 'height.txt')

xs = np.loadtxt(path)
print(xs.shape)

plt.hist(xs, bins='auto', density=True)
plt.xlabel('Height [cm]')
plt.ylabel('Probability Density')
plt.show()
```

```python
mu = np.mean(xs)
sigma = np.std(xs)

print(mu)
print(sigma)

def normal(x, mu=0, sigma=1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-0.5 * ((x - mu) / sigma) ** 2)
    return y


x = np.linspace(150, 190, 1000)
y = normal(x, mu, sigma)

plt.hist(xs, bins='auto', density=True)
plt.plot(x, y)
plt.xlabel('Height [cm]')
plt.ylabel('Probability Density')
plt.show()
```

```python
import os
import numpy as np
import matplotlib.pyplot as plt

#path = os.path.join(os.path.dirname(__file__), 'height.txt')
path = os.path.join(os.getcwd(), 'height.txt')

xs = np.loadtxt(path)

mu = np.mean(xs)
sigma = np.std(xs)
samples = np.random.normal(mu, sigma, 10000)

print(mu)
print(sigma)

plt.hist(xs, bins='auto', density=True, alpha=0.7, label='original')
plt.hist(samples, bins='auto', density=True, alpha=0.7, label='generated')
plt.xlabel('Height [cm]')
plt.ylabel('Probability Density')
plt.legend()
plt.show()
```

```python
import os
import numpy as np
from scipy.stats import norm

path = os.path.join(os.getcwd(), 'height.txt')
xs = np.loadtxt(path)
mu = np.mean(xs)
sigma = np.std(xs)

p1 = norm.cdf(160, mu, sigma)
print('P(X <= 160) =', p1)

p2 = norm.cdf(180, mu, sigma)
print('P(X > 180) =', 1 - p2)
```
