### 지수

```python
x = np.linspace(-5, 5, 801)
y = 2**x
y2 = 0.5**x

y3 = np.exp(x)
y4 = np.log(x)

print(y)
print(y2)

plt.plot(x, y)
plt.show()
plt.plot(x, y2)
plt.show()
plt.plot(x, y3)
plt.show()
plt.plot(x, y4)
plt.show()
```

```python
x = np.linspace(-1, 2, 301)
y = 3**x
y2 = 3**(x-1)+2
graph1, = plt.plot(x, y)
graph2, = plt.plot(x, y2)
plt.legend([graph1, graph2], ["3^x", "3^(x-1)+2"])
plt.show()

x = np.linspace(-1, 2, 301)
y1 = 2**(-x)
y2 = -2**(-x-2)

graph1, = plt.plot(x, y1)
graph2, = plt.plot(x, y2)
plt.legend([graph1, graph2], ["2^(-x)", "-2^(-x-2)"])
plt.show()
```

### 로그함수

```python
import numpy as np
import matplotlib.pyplot as plt

# y1 = log_2(x), y2 = log_2(x1+1)-1
x1 = np.linspace(0.1, 4, 401)
y1 = np.log2(x1)

x2 = np.linspace(-0.9, 4, 501)
y2 = np.log2(x2+1)-1

graph1, = plt.plot(x1, y1)
graph2, = plt.plot(x2, y2)
plt.legend([graph1, graph2], ["log_2(x)", "log_2(x+1)-1"])
plt.show()
```

### 삼각함수

```python
import numpy as np

ang1deg = 60
ang1rad = np.pi/180*ang1deg
print("60도 = ", ang1rad, "rad")
print("60도 = ", np.deg2rad(ang1deg), "rad")

# -3/2pi(rad)
ang2deg = -(3/2)*np.pi
red2 = ang2deg*(180/np.pi)
ang2deg = np.rad2deg(ang2deg)
print("-3/2pi = ", ang2deg, "deg")
print("-3/2pi = ", red2, "deg")
```

```python
# ex 2-14
import numpy as np
import matplotlib.pyplot as plt

# y = sin2(x+1)
x = np.linspace(-1, np.pi-1, 501)
y = np.sin(2*(x+1))
plt.plot(x, y)
plt.show()

# y =-3cosx+1
x = np.linspace(0, 2*np.pi, 501)
y = -3*np.cos(x)+1
plt.plot(x, y)
plt.show()
```

```python
# ex 2-15
import numpy as np
import matplotlib.pyplot as plt

# y=tan2(x-pi/2)
x = np.linspace(-0.25*np.pi, 0.25*np.pi, 501)
y = np.tan(2*(x-0.5*np.pi))
plt.plot(x[10:491], y[10:491])
plt.show()
```
