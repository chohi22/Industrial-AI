```python
# (a) a=5, d=3, s_10?

N = 10
a = 5
d = 3

sum1 = 0
for n in range(1, N+1):
    sum1 += a + (n-1)*d

# sum2 : sn = n(2a + (n-1)d)/2
sum2 = N*(2*a + (N-1)*d)/2

print("S10 = ", sum1)
print("S10 = ", sum2)
```

```python
# Ex 3-7

# (a) sum_{k=1}^10 (k^2-k)
N = 10

# sum 1: Real Summation
sum1 = 0
for k in range(1, N+1):
    sum1 += k**2 - k

# sum 2 : sn = n(n+1)(2n+1)/6 - n(n+1)/2
sum2 = N*(N+1)*(2*N+1)/6 - N*(N+1)/2
print("-----------(a) -----------")
print("S10 = ", sum1)
print("S10 = ", sum2)

# (b) sum_{k=1}^5 k(k^2+3)
N = 5

# sum 1: Real Summation
sum1 = 0
for k in range(1, N+1):
    sum1 += k*(k**2 + 3)

# sum 2 : sn = (n(n+1)/2)^2 + 3*(n(n+1)/2)
sum2 = (N*(N+1)/2.0)**2 + 3*(N*(N+1)/2.0)
print("-----------(b) -----------")
print("S10 = ", sum1)
print("S10 = ", sum2)
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(-2,2,401)
fx = np.zeros_like(x)
for i in range(0,401):
    if np.abs(x[i]) < 1:
        fx[i] = 0
    elif x[i] == 1:
        fx[i] = 0.5
    elif x[i] == -1:
        fx[i] = -0.5
    else:
        fx[i] = x[i]

plt.plot(x,fx,'.')
plt.show()
```
