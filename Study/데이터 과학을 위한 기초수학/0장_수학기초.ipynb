### 합성함수

```python
# Ex 1-7

import numpy as np

# f(x) = x^3+1
def f(x):
    return x**3 + 1

# g(x) = sqrt(x+2)
def g(x):
    return np.sqrt(x + 2)

# (g o f)(2), (f o g)(2), (g o g)(2)
print("(g o f)(2) = ", g(f(2)))
print("(f o g)(2) = ", f(g(2)))
print("(g o g)(2) = ", g(g(2)))
```

### 역함수

```python
# ex 1-14

import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return 3*x-4

def f_inv(x):
    return (x+4)/3

def identity(x):
    return x

x = np.linspace(-1, 5, 601)
graph, = plt.plot(x, f(x))
graph1, = plt.plot(x, f_inv(x))
graph2, = plt.plot(x, identity(x), '--')
plt.legend([graph, graph1, graph2], ['f(x)', 'f_inv(x)', 'identity'])
plt.ylim(-1, 5)
plt.show()

# (b) g(x) = x^2+1
def g(x):
    return x**2 + 1

def g_inv(x):
    return np.sqrt(x-1)

# g = g ^ -1
x = np.linspace(0, 6, 601)
graph, = plt.plot(x, g(x))
graph1, = plt.plot(x, g_inv(x))
graph2, = plt.plot(x, identity(x), '--')
plt.legend([graph, graph1, graph2], ['g(x)', 'g_inv(x)', 'identity'])
plt.ylim(0, 6)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    if x >= -1 and x < 1:
        return -x**2+2
    if x>= 1 and x <=3:
        return x**2-4*x+3
    if x > 3 and x <= 5:
        return -x**2+8*x-14


x = np.linspace(-1, 5, 601)
fx = np.zeros_like(x)


for i in range(0, 601):
    fx[i] = f(x[i])

plt.plot(x, fx, '.')
plt.show()
```
