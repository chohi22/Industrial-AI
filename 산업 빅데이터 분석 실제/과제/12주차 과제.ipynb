## Object type 엔코딩 변환후 예측 결과 올리기

```python
import pandas as pd
df = pd.read_csv('co2.csv')
df.head(5)
```

```python
# object-type columns
object_columns = df.select_dtypes(include=['object']).columns
object_columns
```

```python

object_columns = ['Transmission']

encoded_data = pd.get_dummies(df, columns=object_columns, drop_first=True)
encoded_data
```

```python
import ace_tools_open as tools

tools.display_dataframe_to_user(name="One-Hot Encoded CO2 Dataset", dataframe=encoded_data)

# Display the updated dataframe
encoded_data.head()
```

### Splitting into training and testing datasets (MLR)

```python
encoded_data.columns
```

```python
from sklearn.model_selection import train_test_split

X1_var = encoded_data[[
    'Transmission_A4',
       'Transmission_A5', 'Transmission_A6', 'Transmission_A7',
       'Transmission_A8', 'Transmission_A9', 'Transmission_AM5',
       'Transmission_AM6', 'Transmission_AM7', 'Transmission_AM8',
       'Transmission_AM9', 'Transmission_AS10', 'Transmission_AS4',
       'Transmission_AS5', 'Transmission_AS6', 'Transmission_AS7',
       'Transmission_AS8', 'Transmission_AS9', 'Transmission_AV',
       'Transmission_AV10', 'Transmission_AV6', 'Transmission_AV7',
       'Transmission_AV8', 'Transmission_M5', 'Transmission_M6',
       'Transmission_M7']]
y_var = encoded_data['CO2 Emissions(g/km)'] # dependent variable
X_train, X_test, y_train, y_test = train_test_split(
X1_var,
y_var,
test_size = 0.3,
random_state = 0)
```

### Training model and checking out accuracy (MLR)

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
yhat = lr.predict(X_test)
```

### Checking accuracy (MLR)

```python
from termcolor import colored as cl
print(cl('R-Squared :', attrs = ['bold']),
lr.score(X_test, y_test))
```
