## Clustering

### K-Means in Python

```python
from sklearn.cluster import KMeans
import numpy as np


X = np.array([[2,10], [2,5], [8,4], [5,8], [7,5], [6,4], [1,2], [4,9]])
kmeans = KMeans(n_init=1, n_clusters = 3).fit(X)
print("labels: ", kmeans.labels_)
print("culster centers:", kmeans.cluster_centers_)
print("Predict Values: ", kmeans.predict([[1, 1]]))

```

### K-Means visualization in Python

```python
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.pyplot as plt

X = np.array([[2, 10], [2, 5], [8, 4],[5, 8], [7, 5], [6, 4], [1, 2], [4, 9], [3,3], [4,5], [2,3],[4,5],[6,6], [7,7],
              [3, 11], [1, 4], [3, 2],[1, 2], [7, 5], [4, 2], [2, 1], [3, 7], [4,3], [2,3], [6,3],[6,5],[7,6], [8,7]])
kmeans = KMeans(n_init=3, n_clusters=3).fit(X)
plt.scatter(X[:,0], X[:,1], c=kmeans.labels_, cmap='rainbow')
plt.scatter(kmeans.cluster_centers_[:,0] ,kmeans.cluster_centers_[:,1], color='black', marker="+", s=200)
plt.show()
```

### spiral dataset (spiral.csv) -> n_clusters=3

```python
import pandas as pd

# Load the provided CSV file to check its contents
file_path = 'spiral.csv'
spiral_data = pd.read_csv(file_path)

# Display the first few rows to understand the structure of the data
spiral_data.head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Prepare the data for clustering (only X and Y columns)
X_data = spiral_data[['X', 'Y']]

# Apply K-Means clustering with 3 clusters (since there are 3 groups in the original data)
kmeans = KMeans(n_init=3, n_clusters=3, random_state=42)
spiral_data['KMeans_Cluster'] = kmeans.fit_predict(X_data)

# Plot the original data groups and the KMeans clusters for comparison
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

# Original groups
ax[0].scatter(spiral_data['X'], spiral_data['Y'], c=spiral_data['Group'], cmap='rainbow', s=50)
ax[0].set_title('Original Groups')
ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

# KMeans clusters
ax[1].scatter(spiral_data['X'], spiral_data['Y'], c=spiral_data['KMeans_Cluster'], cmap='rainbow', s=50)
ax[1].set_title('K-Means Clusters')
ax[1].set_xlabel('X')
ax[1].set_ylabel('Y')

plt.tight_layout()
plt.show()
```
