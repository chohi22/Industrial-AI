### 샘플모델 테스트

```python
import numpy as np

def sample(dices=2):
    x = 0
    for _ in range(dices):
        x += np.random.choice([1,2,3,4,5,6])
    return x

print(sample())

print(sample())

print(sample())

print(sample())
```

### 몬테카를로

```python
trial = 1000
samples = []

for _ in range(trial):
    s = sample()
    samples.append(s)


V = sum(samples) / len(samples)

print(V)
```

```python
trial = 1000
samples = []

for _ in range(trial):
    s = sample()
    samples.append(s)
    V = sum(samples) / len(samples)
    print(V)
```

```python
trial = 1000

V, n = 0, 0

for _ in range(trial):
    s = sample()
    n += 1
    V += (s-V) / n #
    print(V)
```
