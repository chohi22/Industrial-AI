### 정규분포

```python
import numpy as np

def normal(x, mu=0, sigma=1):
    return (1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-(x - mu)**2 / (2 * sigma ** 2))
```

```python
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = normal(x)

print(x)
print(y)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

y0 = normal(x, mu=-3)
y1 = normal(x, mu=0)
y2 = normal(x, mu=5)

plt.plot(x, y0, label='$\mu$=-3')
plt.plot(x, y1, label='$\mu$=0')
plt.plot(x, y2, label='$\mu$=5')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

y0 = normal(x, mu=0, sigma=0.5)
y1 = normal(x, mu=0, sigma=1)
y2 = normal(x, mu=0, sigma=2)

plt.plot(x, y0, label='$\sigma$=0.5')
plt.plot(x, y1, label='$\sigma$=1')
plt.plot(x, y2, label='$\sigma$=2')
plt.legend()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_means = []
N = [1,2,4,10,20,50,100,200,500,1000]

for n1 in N:
    x_means = []
    for _ in range(10000):
        xs = []
        for n in range(n1):
            x = np.random.rand()
            xs.append(x)
        mean = np.mean(xs)
        x_means.append(mean)

    # plt.hist(x_means, bins='auto', density=True)
    # plt.title(f'N={n1}')
    # plt.xlabel('x')
    # plt.ylabel('Probability Density')
    # plt.xlim(-0.05, 1.05)
    # plt.ylim(0,5)
    # plt.show()

    # 히스토그램 생성 (normalized to probability density)
    bins = 'auto'
    hist_values, bin_edges, _ = plt.hist(x_means, bins=bins, density=True, alpha=0.6, color='b', label='Histogram')

    # 정규분포 커브 계산
    mean = np.mean(x_means)
    std_dev = np.std(x_means)
    x = np.linspace(min(bin_edges), max(bin_edges), 1000)
    y = (1 / (np.sqrt(2 * np.pi * std_dev ** 2))) * np.exp(-(x - mean) ** 2 / (2 * std_dev ** 2))

    # 정규분포 커브 추가
    plt.plot(x, y, 'r-', label='Normal Curve')

    # 그래프 라벨 및 범위 설정
    plt.title(f'N={n1}')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.xlim(-0.05, 1.05)  # x축은 [0, 1] 범위로 고정
    plt.ylim(0, max(max(hist_values), max(y)) * 1.1)  # y축 자동 조정
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_sums = []
N = 5

for _ in range(10000):
    xs = []
    for n in range(N):
        x = np.random.rand()
        xs.append(x)
    sum = np.sum(xs)
    x_sums.append(sum)

#정규분포
def normal(x, mu=0, sigma=1):
    y = 1 / (np.sqrt(2 * np.pi) * sigma) * np.exp(-(x - mu)**2 / (2 * sigma ** 2))
    return y

x_normal = np.linspace(-5, 5, 1000)
mu = N/2
sigma = np.sqrt(N/12)
y_normal = normal(x_normal, mu=mu, sigma=sigma)

plt.hist(x_sums, bins='auto', density=True)
plt.plot(x_normal, y_normal, 'r-', label='Normal Curve')
plt.title(f'N={N}')
plt.xlim(-1, 6)
plt.show()
```
